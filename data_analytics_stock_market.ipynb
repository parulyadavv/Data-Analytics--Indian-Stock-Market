### import the libraries

```python
import tabula
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

### Read PDF File

```python
#this contain a list
file_path = "./Data_Analytics_-Indian_Stock_Market_Top_2000.pdf"
#Convert your file
df = tabula.read_pdf(file_path, pages = 'all')

```

#### open csv file

```python
DA = pd.read_csv("data1.csv")
DA
```

#### Checking datatypes

```python
DA.dtypes
```

### preprocessing the column 

```python
le = preprocessing.LabelEncoder()

DA['Sub-Sector'] = le.fit_transform(DA['Sub-Sector'])
    
```

```python
DA.head(50)
```

### TASK 1 - screen the companies and visualise them according to sub-sectors using a pie chart
### or a bar charts.

```python
plt.figure(figsize=(40,40))
plt.pie(DA['Sub-Sector'], labels = DA['Ticker'])
plt.axis('equal')
plt.title('screening companies based on sub-sectors')
plt.show()
```

### Renaming the column

```python
DA1 = DA.rename({'Market Cap': 'Market_Cap'}, axis=1)
```

### Data type conversion

```python
DA1['Market_Cap'] = DA1['Market_Cap'].astype('int64')
```

```python
DA1['Market_Cap1'] = pd.cut(DA1['Market_Cap'],bins=[0,50000,200000,2000000], labels = ['small','mid','large'])
```

```python
DA1
```

### TASK 2 -  visualise the companies by segregating them according to Market Cap in
### three categories:

```python
fig = plt.figure(figsize = (40,20))
plt.bar(DA1['Market_Cap1'], DA1['Ticker'], color = 'maroon')
```
